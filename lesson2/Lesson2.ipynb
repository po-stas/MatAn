1. Как относятся друг к другу множество и последовательность? (в ответе использовать слова типа: часть, целое, общее, частное, родитель, дочерний субъект и т.д.)

Последовательность - это множество пронумерованных элементов, где каждому элементу множества сопоставлено натуральное число. Для любого элемента последовательности можно записат следующий за ним элемент. По сути, насколько я понимаю - последовательность - это проиндексированное множество.

2. Прочитать высказывания математической логики, построить их отрицания и установить истинность.



1. Для любого Y от 0 до 1 значение функции сигнум для которого будет равно 1 - в такое множество попадают все значения от 0 до 1 кроме собственно нуля. Поэтому мне кажется высказывание - ложно

Отрицание: 
$\exists y\in[0;1]:sgn(y)\neq1$

2. Для любого n натурального большего чем 2 найдутся x, y, z натуральные, такие что, $x^n = y^n + z^n$
Истинность устанавливал путем разглядывания таблицы квадратов натуральных чисел - действительно возможны такие сочетания скажем x=25, y=20, z=15

Отрицание: Не очень понятно, что мы тут хотим отрицать.. Что не существует $n\in\mathbb N > 2$ для которого ни найдется ни одной трийки $x, y, z\in\mathbb N$ чтобы было верно $x^n = y^n + z^n$ (мне видится что такого n не существует на самом деле, исходя из сделанного выше заключения об истинности)

В таком случае будет: $\nexists n\in\mathbb N > 2 : \nexists x, y, z\in\mathbb N : x^n=y^n+z^n$
Но не факт, что я правильно понимаю суть задачи )

3. Для любого рационального x, найдется рациональное X больше чем x. Мне кажется истинно.

Отрицание: $\nexists x\in\mathbb R \nexists X\in\mathbb R : X > x $

4. Для любого комплексного x не найдется ни одного комплексного y чтобы y было меньше x или y было больше x...
Я плохо знаком с комплексными числами, но думается что высказывание ложно. Поскольку должны быть числа либо большие либо меньшие нашего x

Отрицание:
    $\forall x\in\mathbb C \exists y\in\mathbb C : x > y || x < y$

5. Для любого y в диапазоне [0; $\pi/2$] найдется $\epsilon$ большее чем 0, такое что sin(y) будет меньше чем sin(y + $\epsilon$) - думаю, высказывание ложно - посольку sin($\pi/2$) = 1 и это максимальное значение для функции синуса. Больше он ни при каких значениях аргумента быть не может.

Отрицание:

$\forall y\in[0;\pi/2]\nexists\epsilon>0 : sin(y)<sin(y + \epsilon)$

6. Для любого y в диапазоне $[0;\pi)$ найдется $\epsilon$ большее чем 0 при котором cos(y) будет больше cos(y + $\epsilon$) - мне кажется тоже ложно - поскольку даже не смотря на то, что $\pi$ не включено, cos(0) тоже = 1.. 

Отрицание:
$\forall y\in[0;\pi]\nexists\epsilon > 0 : cos(y) > cos(y+\epsilon)$

7. Существует число x не входящее в объединение всех известных числовых множеств (насколько я понял) - видимо ложно.

Отрицание:
    $\nexists x: x\notin \{\mathbb N,\mathbb Z,\mathbb Q,\mathbb R,\mathbb C\}$
