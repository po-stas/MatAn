–ö–∞–∫ –∏ –≤ –ø—Ä–∏–º–µ—Ä–µ –Ω–∞ –≤–µ–±–∏–Ω–∞—Ä–µ –≤—ã—á–∏—Å–ª—è–µ–º: $N(10^{-7}) = -log_2(10^{-7}) = 23.2534966642$ —Ç–æ –µ—Å—Ç—å 23-–π —ç–ª–µ–º–µ–Ω—Ç –¥–∞—Å—Ç –≤ –ø—Ä–µ–¥–µ–ª–µ —Ç–æ—á–Ω–æ—Å—Ç—å –±–æ–ª—å—à–µ —á–µ–º $10^{-7}$ –º–æ–∂–µ–º –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —á–∏—Å–ª–µ–Ω–Ω–æ:

```python
import math

a_n = 1.0
a_n1 = 0.0
n = 1
e = 0.0000001

while abs(a_n - a_n1) > e:
    a_n = a_n1
    a_n1 += math.sin(n)/pow(2, n)
    n += 1
    
print(f'n = {n}')
print(f'e = {abs(a_n - a_n1)}')
```

# 14*. –ü–æ–ª—å–∑—É—è—Å—å –∫—Ä–∏—Ç–µ—Ä–∏–µ–º –ö–æ—à–∏, –¥–æ–∫–∞–∂–∏—Ç–µ —Ä–∞—Å—Ö–æ–¥–∏–º–æ—Å—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏:

$$b_n = 1 + \frac{1}{2} + \frac{1}{3} + \dots + \frac{1}{n}$$

–õ–∏—á–Ω–æ –º–Ω–µ –Ω–µ –æ—á–µ–Ω—å –ø–æ–Ω—è—Ç–Ω–æ –∑–∞—á–µ–º —Ç—É—Ç –∫—Ä–∏—Ç–µ—Ä–∏–π –ö–æ—à–∏, –≥–∞—Ä–º–æ–Ω–∏—á–µ—Å–∫–∏–π —Ä—è–¥ $\frac{1}{n}$ —Ä–∞—Å—Ö–æ–¥–∏—Ç—Å—è - —Ç–æ–µ—Å—Ç—å —á–∞—Å—Ç–∏—á–Ω—ã–µ —Å—É–º–º—ã —Ä—è–¥–∞ –±—É–¥—É—Ç —Ä–∞—Å—Ç–∏ –¥–æ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ—Å—Ç–∏ –ø—Ä–∏ $n\to\infty$ - –∞ —Ä–∞–∑ –Ω–µ—Ç –ø—Ä–µ–¥–µ–ª–∞ –¥–ª—è —á–ª–µ–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ - —Ç–æ –Ω–µ—Ç –ø—Ä–µ–¥–µ–ª–∞ –∏ —É —Å–∞–º–æ–π –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏.

–ù–æ –µ—Å–ª–∏ –ø–æ–ø—Ä–æ–±–æ–≤–∞—Ç—å –¥–µ–π—Å—Ç–≤–æ–≤–∞—Ç—å —Ç–∞–∫–∏–º –∂–µ –º–µ—Ç–æ–¥–æ–º –∫–∞–∫ –±—ã–ª–æ –ø—Ä–æ–¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä–æ–≤–∞–Ω–æ –Ω–∞ –≤–µ–±–∏–Ω–∞—Ä–µ (–ø—Ä–∏–º–µ—Ä–µ–Ω–∏–µ –∫—Ä–∏—Ç–µ—Ä–∏—è –ö–æ—à–∏) - —Ç–æ –º—ã –ø–æ –ª–æ–≥–∏–∫–µ, –¥–æ–ª–∂–Ω—ã –Ω–∞—Ç–æ–ª–∫–Ω—É—Ç—å—Å—è –Ω–∞ –∫–∞–∫—É—é-–ª–∏–±–æ –Ω–µ–ª–æ–≥–∏—á–Ω–æ—Å—Ç—å - —á—Ç–æ –∏ –¥–æ–ª–∂–Ω–æ –±—É–¥–µ—Ç –ø–æ–∫–∞–∑–∞—Ç—å –Ω–∞–º, —á—Ç–æ –∫—Ä–∏—Ç–µ—Ä–∏–π –Ω–µ —Ä–∞–±–æ—Ç–∞–µ—Ç.. –ß–µ—Å—Ç–Ω–æ –≥–æ–≤–æ—Ä—è –ø–æ–Ω—è—Ç—å - –∫–∞–∫–æ–≥–æ —Ä–æ–¥–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –Ω–µ –ª–æ–≥–∏—á–Ω–æ—Å—Ç—å –º–Ω–µ –Ω–µ —É–¥–∞–ª–æ—Å—å. –ò—Ç–∞–∫: –±–µ—Ä–µ–º $|a_n - a_{n + k}|$

$
\begin{vmatrix}
\frac{1}{2} + \frac{1}{3} + \frac{1}{4} + \dots + \frac{1}{n} - \\
\frac{1}{2} + \frac{1}{3} + \frac{1}{4} + \dots + \frac{1}{n} + \frac{1}{n + k}\\
\end{vmatrix} = 
\begin{vmatrix}
- \frac{1}{n + k}\\
\end{vmatrix} = 
\frac{1}{n + k}
$

–≠—Ç–æ —Ç–æ—á–Ω–æ –º–µ–Ω—å—à–µ —á–µ–º $\frac{1}{n}$ –∏ —Ä–∞—Å—Å—É–∂–¥–∞—è –¥–∞–ª–µ–µ: $\frac{1}{n} < \frac{1}{N(\epsilon)}$

—Ç–æ-–µ—Å—Ç—å $\epsilon = \frac{1}{N(\epsilon)}$

–ó–Ω–∞—á–∏—Ç $N(\epsilon) = \frac{1}{\epsilon}$

–ò –≤–æ—Ç —Ç—É—Ç —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ –≤–æ–ø—Ä–æ—Å: –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç—å –≤—Ä–æ–¥–µ –µ—Å—Ç—å, –º—ã –≤—Å–µ–≥–¥–∞ –º–æ–∂–µ–º –ø–æ–ª—É—á–∏—Ç—å —Ç–æ—Ç –Ω–æ–º–µ—Ä —á–ª–µ–Ω–∞ –≤ —Ä–∞–π–æ–Ω–µ –∫–æ—Ç–æ—Ä–æ–≥–æ –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å –±—É–¥–µ—Ç —Å–æ–±–ª—é–¥–∞—Ç—å—Å—è. –î–ª—è $10^{-3}$ —ç—Ç–æ –±—É–¥–µ—Ç $N(\epsilon) = 1000$, –¥–ª—è $10^{-7} \rightarrow N(\epsilon) = 10000000$, –¥–ª—è $10^{-10} \rightarrow N(\epsilon) = 10000000000$ –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ. –≠—Ç–æ –¥–∞–∂–µ –º–æ–∂–Ω–æ –ø—Ä–æ–≤–µ—Ä–∏—Ç—å —á–∏—Å–ª–µ–Ω–Ω–æ –∫–∞–∫ –∏ –≤ –ø—Ä–µ–¥—ã–¥—É—â–µ–º –ø—Ä–∏–º–µ—Ä–µ. 

```python
a_n = 1.0
a_n1 = 0.0
n = 1
e = 0.000001

while abs(a_n - a_n1) > e:
    a_n = a_n1
    a_n1 += 1.0/n
    n += 1
    
print(f'n = {n}')
print(f'e = {abs(a_n - a_n1)}')
```

–ê –Ω–∞–º –ø–æ–∏–¥–µ–µ –Ω—É–∂–Ω–æ, —á—Ç–æ–±—ã –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç—å —Å–æ–±–ª—é–¥–∞–ª–∞—Å—å –Ω–µ —Ç–æ–ª—å–∫–æ "–≤ —ç—Ç–æ–º —á–ª–µ–Ω–µ" - –Ω–æ –∏ –≤–µ–∑–¥–µ –¥–∞–ª—å—à–µ. –ê –≤–æ—Ç —Å —ç—Ç–∏–º - –ø—Ä–æ–±–ª–µ–º–∞. –ï—Å–ª–∏ –æ—Ç—Å—Ç—É–ø–∞—Ç—å –æ—Ç –Ω–∞—à–µ–≥–æ $N(\epsilon)$ –¥–∞–ª—å—à–µ —á–µ–º –Ω–∞ –æ–¥–∏–Ω —á–ª–µ–Ω - —Ç–æ –≤–∏–¥–Ω–æ, —á—Ç–æ —Ä–∞–∑–Ω–æ—Å—Ç—å —Ä–∞—Å—Ç–µ—Ç –ª–æ–º–∞—è –≤–µ—Å—å –∫—Ä–∏—Ç–µ—Ä–∏–π.

```python
a_n = 0.0
a_n1000 = 0.0
for n in range(1, 1100):
    a_n += 1.0/n
    if n == 1000:
        a_n1000 = a_n
    if n > 1000:
        print(f'{abs(a_n - a_n1000)}')

```

–ï—Å–ª–∏ –≤–∑—è—Ç—å –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å —Ä—è–¥–æ–≤ –∏–∑ –ø—Ä–µ–¥—ã–¥—É—â–µ–≥–æ –∑–∞–¥–∞–Ω–∏—è - —Ç–æ —Ç—É—Ç –≤–∏–¥–Ω–æ, —á—Ç–æ –≤—Å–µ –æ–∫. –¢–æ—á–Ω–æ—Å—Ç—å –Ω–µ –ø—Ä–µ–≤—ã—à–∞–µ—Ç—Å—è –¥–ª—è –≤—Å–µ—Ö —á–ª–µ–Ω–æ–≤ –±–æ–ª—å—à–µ –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–≥–æ.

```python
a_n = 0.0
a_n23 = 0.0
for n in range(1, 50):
    a_n += math.sin(n)/pow(2, n)
    if n == 23:
        a_n23 = a_n
    if n > 23:
        print(f'{abs(a_n - a_n23)}')
```

–û—Å—Ç–∞–µ—Ç—Å—è –≤–æ–ø—Ä–æ—Å - –∫–∞–∫ –º—ã –¥–æ–ª–∂–Ω—ã –ø–æ–Ω—è—Ç—å, —á—Ç–æ –ø–æ–ª—É—á–µ–Ω–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è $N(\epsilon)$ –Ω–µ –æ–±–µ—Å–ø–µ—á–∏–≤–∞–µ—Ç —Å—Ö–æ–∂–¥–µ–Ω–∏—è? –ò–ª–∏ —è —Å–æ–≤—Å–µ–º –Ω–∏—á–µ–≥–æ –Ω–µ –ø–æ–Ω—è–ª –∏–∑ –ø—Ä–∏–º–µ—Ä–∞ –Ω–∞ –≤–µ–±–∏–Ω–∞—Ä–µ?
