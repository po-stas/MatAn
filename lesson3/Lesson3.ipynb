# 13. Пользуясь критерием Коши, докажите сходимость последовательности:

$$a_n = \frac{sin(1)}{2} + \frac{sin(2)}{2^2} + \frac{sin(3)}{2^3} + \dots + \frac{sin(n)}{2^n}$$

Применяя рассмотренный на лекции метод применения критерия Коши, возьмем два члена последовательности n-ный и n+1-ый. И вычтем один из другого

$
\begin{vmatrix}
\frac{sin(1)}{2} + \frac{sin(2)}{2^2} + \frac{sin(3)}{2^3} + \dots + \frac{sin(n)}{2^n} - \\
\frac{sin(1)}{2} + \frac{sin(2)}{2^2} + \frac{sin(3)}{2^3} + \dots + \frac{sin(n)}{2^n} + \frac{sin(n + 1)}{2^{n + 1}}\\
\end{vmatrix} = 
\begin{vmatrix}
- \frac{sin(n + 1)}{2^{n + 1}}\\
\end{vmatrix} = 
\frac{sin(n + 1)}{2^{n + 1}}
$

Это должно быть меньше $\epsilon$

Представим себе, что может быть точно больше этого члена - на мой взгляд $\frac{sin(n + 1)}{2^{n + 1}} < \frac{1}{2^n}$ поскольку $sin$ чего бы то ни было не может быть больше 1, а в знаменателе меньшая степень - даст в итоге большее общее значение дроби (по крайней мере я не вижу логических причин чтобы так нельзя было рассуждать)

Соответственно в точности как и в примере на вебинаре $\frac{1}{2^{N(\epsilon)}} = \epsilon$, то-есть критерий выполняется. Последовательность сходится.

# * Какой член последовательности можно взять в качестве предела с точностью 𝜀 = 10−7

Как и в примере на вебинаре вычисляем: $N(10^{-7}) = -log_2(10^{-7}) = 23.2534966642$ то есть 23-й элемент даст в пределе точность больше чем $10^{-7}$ можем проверить численно:

```python
import math

a_n = 1.0
a_n1 = 0.0
n = 1
e = 0.0000001

while abs(a_n - a_n1) > e:
    a_n = a_n1
    a_n1 += math.sin(n)/pow(2, n)
    n += 1
    
print(f'n = {n}')
print(f'e = {abs(a_n - a_n1)}')
```

Но если попробовать действовать таким же методом как было продемонстрировано на вебинаре (примерение критерия Коши) - то мы по логике, должны натолкнуться на какую-либо нелогичность - что и должно будет показать нам, что критерий не работает.. Честно говоря понять - какого рода должна быть не логичность мне не удалось. Итак: берем $|a_n - a_{n + k}|$

$
\begin{vmatrix}
\frac{1}{2} + \frac{1}{3} + \frac{1}{4} + \dots + \frac{1}{n} - \\
\frac{1}{2} + \frac{1}{3} + \frac{1}{4} + \dots + \frac{1}{n} + \frac{1}{n + k}\\
\end{vmatrix} = 
\begin{vmatrix}
- \frac{1}{n + k}\\
\end{vmatrix} = 
\frac{1}{n + k}
$

Это точно меньше чем $\frac{1}{n}$ и рассуждая далее: $\frac{1}{n} < \frac{1}{N(\epsilon)}$

то-есть $\epsilon = \frac{1}{N(\epsilon)}$

Значит $N(\epsilon) = \frac{1}{\epsilon}$

И вот тут собственно вопрос: зависимость вроде есть, мы всегда можем получить тот номер члена в районе которого погрешность будет соблюдаться. Для $10^{-3}$ это будет $N(\epsilon) = 1000$, для $10^{-7} \rightarrow N(\epsilon) = 10000000$, для $10^{-10} \rightarrow N(\epsilon) = 10000000000$ и так далее. Это даже можно проверить численно как и в предыдущем примере. 

```python
a_n = 1.0
a_n1 = 0.0
n = 1
e = 0.000001

while abs(a_n - a_n1) > e:
    a_n = a_n1
    a_n1 += 1.0/n
    n += 1
    
print(f'n = {n}')
print(f'e = {abs(a_n - a_n1)}')
```

А нам поидее нужно, чтобы погрешность соблюдалась не только "в этом члене" - но и везде дальше. А вот с этим - проблема. Если отступать от нашего $N(\epsilon)$ дальше чем на один член - то видно, что разность растет ломая весь критерий.

```python
a_n = 0.0
a_n1000 = 0.0
for n in range(1, 1100):
    a_n += 1.0/n
    if n == 1000:
        a_n1000 = a_n
    if n > 1000:
        print(f'{abs(a_n - a_n1000)}')

```
